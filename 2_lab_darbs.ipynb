```python
import numpy as n
import matplotlib.pyplot as p
```

```python
x = n.arange(0, 4, 0.01)
y_cos = n.cos(x)
y_sin = n.sin(x)
f1 = x
f2 = x - n.power(x, 3)/n.math.factorial(3)
f3 = x - n.power(x, 3)/n.math.factorial(3) + n.power(x, 5)/n.math.factorial(5)
f4 = x - n.power(x, 3)/n.math.factorial(3) + n.power(x, 5)/n.math.factorial(5) - n.power(x,7)/n.math.factorial(7)
```

```python
p.title('Funkcijas $sin(x), cos(x)$')
p.xlabel('x')
p.ylabel('f(x)')
p.grid()
p.plot(x,y_cos, color = "#ff0000")

```
