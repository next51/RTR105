```python
import math as m
import numpy as n
from time import sleep
import matplotlib.pyplot as p
```

```python
def f(x):
    return n.power(n.cos(x),2)
a = 1. * float(input('a robeza: '))
b = 1. * float(input('b robeza: '))
```

```python
funa = f(a)
funb = f(b)
```

```python
deltax = 0.0001
x = n.arange(a, b, 0.01)
p.title("cos^2(x) funkcija")
p.plot(x, f(x))
p.show()
```

```python
while (m.fabs(b-a) > deltax):
    x= (a+b)/2
    funx = f(x)
if (funa*funx < 0.):
    b=x
else:
    a=x
print("Sakne ir:",x)
```
